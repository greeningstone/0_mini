### 그럼 성별로 jop에 관련된 변수가 어떻게 달라지는지 확인해보자

```
job_ratio_sex = jeju[['job_majorc','job_smallc','job_public','job_profession','job_self','job_none','job_other']].groupby([jeju.sex]).mean().reset_index()
```

```
job_ratio_sex_table = job_ratio_sex.T.reset_index()
job_ratio_sex_table.columns = ['job','male','female']
```

```
job_ratio_sex_table.iloc[1:,:]
```

```
job_ratio_sex_final = pd.melt(job_ratio_sex_table.iloc[1:,:],id_vars = 'job', var_name = 'sex', value_name = 'job_ratio')
```

```
job_ratio_sex_final
```

```
sns.barplot(x='job',y='job_ratio',hue='sex',data = job_ratio_sex_final) #평균
```

### 그럼 이제 성별, 나이별로 어떻게 job ratip들이 달라지는지 보자

```
job_ratio_sex_age = jeju[['job_majorc','job_smallc','job_public','job_profession','job_self','job_none','job_other']].groupby([jeju.sex, jeju.age]).mean().reset_index()
```

```
job_ratio_sex_age.sum(axis=0)
```

```
job_ratio_sex_age
```

```
job_ratio_sex_age[['job_majorc','job_smallc','job_public','job_profession','job_self','job_none','job_other']][(job_ratio_sex_age.sex==2)&(job_ratio_sex_age.age==24)].sum(axis=0).sum()
```

```
job_ratio_sex_age_table = pd.melt(job_ratio_sex_age, id_vars=['sex','age'], value_vars = ['job_majorc','job_smallc','job_public','job_profession','job_self','job_none','job_other'], value_name = 'job_ratio')
```

```
grid = sns.FacetGrid(job_ratio_sex_age_table, col='variable',hue='sex', palette = 'RdBu_r', col_wrap=4, height=3,legend_out=True)

grid.map(plt.plot, 'age', 'job_ratio', marker='o')

grid.set(yticks=[0, 1])

grid.fig.tight_layout(w_pad=1)
```
